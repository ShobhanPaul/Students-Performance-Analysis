```python
import pandas as pd
df = pd.read_csv('student_habits_performance.csv')
df.head()
```

```python
import seaborn as sns
sns.kdeplot(df['age'],bw_adjust=1)
sns.despine()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
df['age']=pd.to_numeric(df['age'],errors='coerce')
df.replace([np.inf,-np.inf],np.nan,inplace=True)
plt.figure(figsize=(10,10))
sns.kdeplot(df['age'].dropna(),bw_adjust=1,label='bw_adjust_1',color='red')
sns.kdeplot(df['age'].dropna(),bw_adjust=0.5,label='bw_adjust_0.5',color='blue')
plt.legend(loc=0)
```

```python
ax = sns.countplot(x='gender',hue='part_time_job',data=df)
for i in ax.containers:
    ax.bar_label(i,padding=3)
```
